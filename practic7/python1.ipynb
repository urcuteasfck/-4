```python
def findMinElement(c):
    k = int(c[0])
    for i in range(len(c)):
        if int(c[i])<k:
            k = int(c[i])
    return k

def ArrDivLowestElement(a,b):
    minA = findMinElement(a)
    minB = findMinElement(b)
    c = 0
    z = 0
    print(*a)
    print(*b)
    if minA!=0:
        for i in range(len(a)):
            a[i] = int(a[i])//minA
        print(*a)
    else:
        print('Ошибка - деление на 0')
    if minB!=0:
        for i in range(len(b)):
            b[i] == int(b[i]) // minB
        print(*b)
    else:
        print('Ошибка - деление на 0')
    return ''
a = input().split()
b = input().split()
print(ArrDivLowestElement(a,b))
```
