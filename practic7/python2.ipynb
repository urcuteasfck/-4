```python
def set_diff(a,b):
    a_b=a-b
    zipp = []
    for i in a_b:
        if int(i)>=-5 and int(i)<=5:
            zipp.append(i)
    return zipp
setA = set(input().split())
setB =  set(input().split())
print(set_diff(setA,setB))
```
